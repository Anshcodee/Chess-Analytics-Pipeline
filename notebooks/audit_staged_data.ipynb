Pathing and Loading

```python
# Assuming this notebook is inside a 'notebooks/' folder at the project root
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))

games_file = os.path.join(PROJECT_ROOT, 'data', '02_staged', 'games', 'games_0001.parquet')
moves_file = os.path.join(PROJECT_ROOT, 'data', '02_staged', 'moves', 'moves_0001.parquet')

print(f"Loading from: {games_file}")
df_games = pd.read_parquet(games_file)
df_moves = pd.read_parquet(moves_file)
print("Data loaded successfully!")
```

Testing

```python
import os
import glob
import pandas as pd

# 1. Build an absolute path dynamically (works regardless of where the notebook is run)
base_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
file_pattern = os.path.join(base_dir, 'data', '02_staged', 'moves', '*.parquet')

# 2. Grab the files
files = glob.glob(file_pattern)
print(f"Found {len(files)} files.")

# 3. Safely concatenate
if files:
    df = pd.concat(pd.read_parquet(f) for f in files)
    print(df['color'].value_counts(normalize=True))
else:
    print(f"Failed to find files. Checked path: {file_pattern}")
```
